## Data Visualization

```python
# Chage several data {Attrition,Over18,OverTime} into 1 or 0 (numeric boolean)
DATA_FRAME['Attrition'] = DATA_FRAME['Attrition'].apply(lambda x:1 if x == "Yes" else 0)
DATA_FRAME["Over18"] = DATA_FRAME["Over18"].apply(lambda x:1 if x =="Y" else 0)
DATA_FRAME["OverTime"] = DATA_FRAME["OverTime"].apply(lambda x:1 if x == "Yes" else 0)
```

```python
# We're going to drop the unused data ()
DATA_FRAME.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1, inplace= True)
```

```python
DATA_FRAME.head(20)
```

```python
# how mane employee who will left the company !
left_df = DATA_FRAME[DATA_FRAME['Attrition'] == 1]
stayed_df = DATA_FRAME[DATA_FRAME['Attrition'] == 0]
print("Total Data Employee : ", len(DATA_FRAME))
print("Total left : ",len(left_df))
print("Total stayed : ",len(stayed_df))
```

- count the number of employees who stayed and left
- plot the number of employees who left
- plot the number of employees who stayed
- percentace the number of employees who left
- percentace the number of employees who stayed

```python
TotalDataemployee  =  len(DATA_FRAME)
Employeeleftdata   =  len(left_df)
Employeestayeddata = len(stayed_df)
print(TotalDataemployee,Employeeleftdata,Employeestayeddata)
```

```python

def countPercentage(currentScore, totalScore):
    X = currentScore / totalScore * 100
    return X
```

```python
# count the number of employees who stayed and left
whoLeft = left_df['Attrition'].count()
print("number of employee who left is : ",whoLeft)
```

```python
# plot the number of employees who stayed
whoStayed = stayed_df['Attrition'].count()
print("number of employee who stayed is : ",whoStayed)
```

```python
X = countPercentage(Employeeleftdata,TotalDataemployee)
Y = countPercentage(Employeestayeddata,TotalDataemployee)
print("number of Employees : " ,TotalDataemployee)
print("Percentage left empoyees : ",X,"%")
print("Percentage Stayed empoyees : ",Y,"%")
```
